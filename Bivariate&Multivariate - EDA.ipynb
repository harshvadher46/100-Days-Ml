```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
df1 = sns.load_dataset('iris')
```

```python
df2 = sns.load_dataset('titanic')
```

```python
df3 = sns.load_dataset('flights')
```

## Scatter Plot(Numerical- Numerical)

```python
sns.scatterplot(df,x='total_bill',y='tip',hue='sex',style='smoker',size='size')
```

## Bar plot (Numerical - Categorical)

```python
# it show that the category vise age group mostly travel
sns.barplot(df2,x='pclass',y='age',hue='sex') 
```

```python
sns.barplot(df2,x='pclass',y='fare',hue='sex')
```

## Boxplot(Numerical - Categorical)

```python
sns.boxplot(df2,x='sex',y='age',hue='survived')
```

## Distplot(Numerical - Categorical)

```python
sns.distplot(df2,x='')
```

## Heatmap(Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(df2['pclass'],df2['survived']))
# The lighter(cream shade shows that the number of people in class 3 were died the most)
```

```python
(df2.groupby('pclass').mean()['survived']*100).plot(kind='bar')
```

```python
df2.groupby('sex').mean()['survived']*100
```

```python
# df2.groupby('embarked').mean()['survived'].plot
# sns.(df2,x='embarked',y='sex')
```

## ClusterMap(Categorical-categorical)

```python
sns.clustermap(pd.crosstab(df2['parch'],df2['survived'])) # The side bar shows that the connected ones are highly related to each other
```

### Pair plot 

```python
# it shows all the relationship of all the (NUMERICAL) columns with each other using the scatter plot
sns.pairplot(df1,hue='species')
```

## LinePlot(Numerical-Numerical)

```python
new = df3.groupby('year').sum().reset_index()
```

```python
new
```

```python
sns.lineplot(new,x='passengers',y='year') # Here you can clearly see that the number of customer grows as the year increases
```
